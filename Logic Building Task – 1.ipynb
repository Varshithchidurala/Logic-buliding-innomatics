```python
2.

marks= [45, 78, 90, 33, 60]

pass_inital=0
fail_inital=0
for m in marks:
     if m >= 50:
         pass_inital=pass_inital+1
     else:
         fail_inital=fail_inital+1

print("Pass:",pass_inital)
print("Fail:",fail_inital)
```

```python
3.

names = [" Alice ", "bob", " CHARLIE "]

clean = []
for name in names:
    name = name.strip()
    name = name.lower()
    clean.append(name)

print("Cleaned Names:", clean)
```
