```python
1.

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == highest_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
2.

query = "Buy mobile phone buy phone online"
query = query.lower()
words = query.split()
count = {}
for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1
result = {}
for word in count:
    if count[word] > 1:
        result[word] = count[word]
print(result)
```

```python
3.

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid = []
for i, value in enumerate(sensor_readings):
    if value % 2 == 0:   # check even
        valid.append((i, value))
print(valid)
```

```python
4.

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

count = {}
for email in emails:
    domain = email.split("@")[1]
    count[domain] = count.get(domain, 0) + 1
total = len(emails)
for domain in count:
    print(domain, ":", int((count[domain] / total) * 100), "%")
```

```python
5.

sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
for i, value in enumerate(sales):
    if value > average * 1.2:
        print(f"Day {i+1}: {value}")
```
