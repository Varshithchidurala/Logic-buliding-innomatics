```python
1.

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return f"Invalid plan. Available plans are: {valid_plans}"
    else:
        return "VALID"
while True:
    try:
        amount = int(input("Enter recharge amount:"))
        result = validate_recharge(amount)
        
        if result == "VALID":
            print("Recharge successful!")
            break   # Exit loop after successful recharge
        else:
            print("invalid", result)
            print("Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```

```python
3.


def check_result(marks):
    total = 0
    for m in marks:
        total = total + m

    average = total / len(marks)
    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"

marks_list = [60, 40, 50, 70]

avg, result = check_result(marks_list)

print("Average:", avg)
print("Result:", result)
```

```python
4.

    
def get_fare(distance, peak):
    fare = 50 + (12 * distance)   
    
    if peak == "yes":
        fare = fare + (fare * 0.25)   
    
    return fare

while True:
    try:
        km = float(input("Enter distance (km): "))
        peak_time = input("Peak hour? (yes/no): ").lower()
        
        total = get_fare(km, peak_time)
        print("Total Fare: ₹", total)
        break
    
    except:
        print("Invalid input. Try again.\n")
```

```python
5.


def check_eligibility(attendance):
    present_days = 0
    
    for day in attendance:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance)) * 100
    
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"

attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_eligibility(attendance_list)

print("Attendance Percentage:", percent)
print("Status:", status)
```

```python
6.


def check_password(pwd):
    if len(pwd) < 8:
        return "Weak Password"
    digit = False
    special = False
    
    for ch in pwd:
        if ch.isdigit():
            digit = True
        if ch in "@#$":
            special = True
    
    if digit and special:
        return "Strong Password"
    else:
        return "Weak Password"
password = input("Enter password: ")
print(check_password(password))
```
